```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Beispielbild einlesen (verwende dein eigenes Bild)
image = cv2.imread('dacl10k_v2_train_0005.jpg', cv2.IMREAD_GRAYSCALE)

# Kanten mit Canny Edge Detection erkennen
edges = cv2.Canny(image, 50, 150, apertureSize=3)

# Hough-Transformation zur Linienerkennung anwenden
lines = cv2.HoughLines(edges, 1, np.pi / 180, 100)

# Bild in BGR umwandeln, um die Linien in Farbe darzustellen
image_with_lines = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

# Linien zeichnen
if lines is not None:
    for rho, theta in lines[:, 0]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        cv2.line(image_with_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

# Bild anzeigen
plt.figure(figsize=(10, 10))
plt.imshow(image_with_lines)
plt.title('Erkannte Linien')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Beispielbild einlesen (verwende deinen eigenen Bildpfad)
image_path = 'dacl10k_v2_train_0005.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Überprüfen, ob das Bild korrekt geladen wurde
if image is None:
    print(f"Fehler: Bild unter dem Pfad {image_path} konnte nicht geladen werden.")
else:
    # Kanten mit Canny Edge Detection erkennen
    edges = cv2.Canny(image, 50, 150, apertureSize=3)

    # Hough-Transformation zur Linienerkennung anwenden
    lines = cv2.HoughLines(edges, 1, np.pi / 180, 100)

    # Bild in BGR umwandeln, um die Linien in Farbe darzustellen
    image_with_lines = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    # Linien zeichnen
    if lines is not None:
        for rho, theta in lines[:, 0]:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a * rho
            y0 = b * rho
            x1 = int(x0 + 1000 * (-b))
            y1 = int(y0 + 1000 * (a))
            x2 = int(x0 - 1000 * (-b))
            y2 = int(y0 - 1000 * (a))
            cv2.line(image_with_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

    # Bild anzeigen
    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image_with_lines, cv2.COLOR_BGR2RGB))
    plt.title('Erkannte Linien')
    plt.axis('off')
    plt.show()
```
